```python
def show_grids(grids):
    fig, axes = plt.subplots(2, len(grids), figsize=(len(grids)*2.5, 5))
    plt.tick_params(left = False, right = False, labelleft = False, labelbottom = False, bottom = False)
    axes = axes.reshape((2, len(grids)))
    for i in range(len(grids)):
        axes[0][i].pcolormesh(grids[i]['input'][::-1])
        axes[1][i].pcolormesh(grids[i]['output'][::-1])
        axes[0][i].set_axis_off()
        axes[1][i].set_axis_off()
    plt.show()
```

### Count number of tokens a message to GPT 3.5 would take

```python
def num_tokens(text, model=GPT_MODEL):
    encoding = tiktoken.encoding_for_model(model)
    return len(encoding.encode(text))
```

### Create message with training inputs and outputs as well as testing inputs to send to GPT 3.5

```python
def construct_message(task):
    token_budget = 4000

    description = \
"""\
\"Given a series of input-output pairs of grids to train on, find the rule that transforms \
input grids to output grids. Then use that rule on the test input grid to produce test \
output grids. Only write the outputs and nothing else. If there are multiple test inputs, \
separate the multiple test outputs with a semicolon ';'. Each test output should be outputed \
as a 2D array, using square brackets and commas to separate. \
"""

    message_template = Template(description + \
"""\
Here are the training pairs:\\n\\n\
$train_pairs\\n\\n\
Here are the testing inputs:\\n\\n\
$testing_inputs\"\
""")

    testing_inputs = ';'.join([str(pair['input']) for pair in task['test']]).replace(' ', '')
    message = message_template.substitute(
        train_pairs='',
        testing_inputs=testing_inputs
    )
    for i in range(len(task['train'])):
        message_with_added_pair = message_template.substitute(
            train_pairs=str(task['train'][0:i+1]).replace(' ', ''),
            testing_inputs=testing_inputs
        )
        if num_tokens(message_with_added_pair) < token_budget:
            message = message_with_added_pair
        
    return message
```

### Create message template for the data to send to OpenAI in order to fine-tune GPT 3.5

```python
message_template = Template("""\
{"messages": [\
{"role": "system", "content": "You learn rules which transform input grids into output grids."},\
{"role": "user", "content": $user_message},\
{"role": "assistant", "content": "$assistant_message"}\
]}""")
```

```python
message_template.substitute(
    user_message = construct_message(task),
    assistant_message = ';'.join([str(pair['output']) for pair in task['test']]).replace(' ', '')
)
```

### Create training and evaluation files to send to OpenAI in order to fine-tune GPT 3.5

```python
def create_jsonl_file(partition):
    with open(f'{partition}_file.jsonl', 'w') as partition_file:
        lines = []
        for file_name in glob.glob(f'./{partition}/*.json'):
            with open(file_name, 'r') as partition_task_file:
                partition_task = json.load(partition_task_file)
                line = message_template.substitute(
                    user_message=construct_message(partition_task),
                    assistant_message=';'.join([str(pair['output']) for pair in partition_task['test']]).replace(' ', '')
                ) + '\n'
                lines.append(line)
        partition_file.writelines(lines)
```

```python
for partition in ['training', 'evaluation']:
    create_jsonl_file(partition)
```

```python
training_file = client.files.create(
    file=open('training_file.jsonl', 'rb'),
    purpose='fine-tune'
)
```

```python
evaluation_file = client.files.create(
    file=open('evaluation_file.jsonl', 'rb'),
    purpose='fine-tune'
)
```

### Fine-tune GPT 3.5 on this data

```python
fine_tuning_job = client.fine_tuning.jobs.create(
    training_file=training_file.id,
    validation_file=evaluation_file.id,
    model=GPT_MODEL,
    suffix='arc'
)
```

```python
# Retrieve the state of a fine-tune
fine_tuning_job_details = client.fine_tuning.jobs.retrieve(fine_tuning_job.id)
```

### Testing the fine-tuned model on the evaluation dataset. The model is unable to generalize and solve most problems. About half of the responses from the model were cut off. This is probably a limitation of GPT 3.5. Additionally, since the model isn't GPT 4, it does have much more limitations with reasoning, and so was not able to successfully answer most of the other questions it did provide full answers for.

```python
correct_count = 0
incorrect_count = 0
error_count = 0
total_count = 0
for idx, file_name in enumerate(glob.glob('./evaluation/*.json')):
    with open(file_name, 'r') as test_task_file:
        test_task = json.load(test_task_file)
        messages = [
            {'role': 'system', 'content': 'You learn rules which transform input grids into output grids.'},
            {'role': 'user', 'content': construct_message(test_task)},
        ]
        overall_response = ''
        for j in range(2):
            response = client.chat.completions.create(
                model=fine_tuning_job_details.fine_tuned_model,
                messages=messages,
                temperature=0
            )
            overall_response += response.choices[0].message.content
            overall_responses = overall_response.split(';')
            error = False
            correct = True
            for i in range(len(overall_responses)):
                try:
                    if eval(overall_responses[i]) != test_task['test'][i]['output']:
                        correct = False
                    show_grids(test_task['test'])
                    show_grids([{
                        'input': test_task['test'][i]['input'],
                        'output': eval(overall_responses[i])
                    }])
                except:
                    print(overall_responses)
                    error = True
                    messages = [
                        {'role': 'system', 'content': 'You learn rules which transform input grids into output grids.'},
                        {'role': 'user', 'content': 
'The array you returned was cutoff. Continue extending it \
from where you left off, making sure that the combined response \
both correct and in the correct format. Only output the array, \
no extra words. Your response is being parsed with code, so this \
is a strict requirement.'
                        },
                    ]
            if not error:
                if correct:
                    correct_count += 1
                else:
                    incorrect_count += 1
                break
            elif j == 1:
                error_count += 1
    total_count = correct_count + incorrect_count + error_count
    print(
f'Completed: {idx+1}/{total_count}\n\
Correct: {correct_count}/{total_count} = {correct_count/total_count}\n\
Incorrect: {incorrect_count}/{total_count} = {incorrect_count/total_count}\n\
Error: {error_count}/{total_count} = {error_count/total_count}\n'
    )
```
